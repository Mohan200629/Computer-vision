```python
import cv2
import numpy as np

def draw_dense_optical_flow(prev_img, next_img, step=10):
    next_img = cv2.resize(next_img, (prev_img.shape[1], prev_img.shape[0]))

    prev_gray = cv2.cvtColor(prev_img, cv2.COLOR_BGR2GRAY)
    next_gray = cv2.cvtColor(next_img, cv2.COLOR_BGR2GRAY)

    flow = cv2.calcOpticalFlowFarneback(prev_gray, next_gray, None, 0.5, 3, 15, 3, 5, 1.2, 0)

    output = next_img.copy()

    h, w = prev_gray.shape
    for y in range(0, h, step):
        for x in range(0, w, step):
            fx, fy = flow[y, x]
            cv2.arrowedLine(output, (x, y), (int(x + fx), int(y + fy)), (0, 0, 255), 1, tipLength=0.3)

    return output


cap = cv2.VideoCapture("E:/murugesan/Downloads/854733-hd_1920_1080_24fps (1).mp4") 

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))

out = cv2.VideoWriter('output_optical_flow.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))


ret, prev_frame = cap.read()

while True:
    ret, next_frame = cap.read()
    if not ret:
        break

    output_img = draw_dense_optical_flow(prev_frame, next_frame)
    cv2.imshow("Dense Optical Flow", output_img)
    out.write(output_img)
    prev_frame = next_frame.copy()
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
